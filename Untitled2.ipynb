```python
import numpy as np

# Parameters
notional = 1000000  # Notional amount of the derivative contract
recovery_rate = 0.4  # Recovery rate in case of default
risk_free_rate = 0.02  # Risk-free interest rate
volatility = 0.2  # Volatility of the underlying asset (e.g., ETH)
correlation = 0.3  # Correlation between the asset and counterparty
maturity = 1  # Maturity of the derivative contract in years
num_simulations = 10000  # Number of Monte Carlo simulations

# Simulate asset values (e.g., ETH prices)
np.random.seed(42)
eth_prices = np.random.normal(0, np.sqrt(maturity) * volatility, num_simulations)

# Simulate counterparty values
counterparty_values = correlation * eth_prices + np.sqrt(1 - correlation**2) * np.random.normal(0, np.sqrt(maturity), num_simulations)

# Calculate the present value of the derivative
discount_factor = np.exp(-risk_free_rate * maturity)
derivative_values = np.maximum(eth_prices - counterparty_values, 0)
present_value = np.mean(derivative_values) * discount_factor * notional

# Calculate the potential future exposure (PFE)
pfe = np.percentile(derivative_values * discount_factor * notional, 95)

# Security deposit in ETH
initial_collateral = 50  # Initial collateral in ETH
collateral_prices = eth_prices * initial_collateral

# Calculate the mark-to-market value of the collateral
mtm_collateral = np.mean(collateral_prices)

# Calculate Credit Valuation Adjustment (CVA) considering collateral
cva = (1 - recovery_rate) * max(pfe - mtm_collateral, 0)

# Display the results
print(f"Present Value of Derivative: {present_value:.2f}")
print(f"Potential Future Exposure (PFE): {pfe:.2f}")
print(f"Mark-to-Market Value of Collateral: {mtm_collateral:.2f}")
print(f"Credit Valuation Adjustment (CVA) with Collateral: {cva:.2f}")
```
